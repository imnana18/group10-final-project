```python
# Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Preview CSV data
data = Path('../Data/clean_fetal_data.csv')
df = pd.read_csv(data)
df.head()
```

```python
# Total rows
total_rows = df.shape[0]
print(total_rows)
```

```python
# Finding max and min percentages (% Time with Abnormal Short-Term Variance) for binning
ST_max_value = df['abnorm_ST_Var_Perc'].max()
ST_min_value = df['abnorm_ST_Var_Perc'].min()

print("Maximum value:", ST_max_value)
print("Minimum value:", ST_min_value)
```

```python
# Finding max and min percentages (% Time with Abnormal Long-Term Variance) for binning
LT_max_value = df['abnorm_LT_Var_Perc'].max()
LT_min_value = df['abnorm_LT_Var_Perc'].min()

print("Maximum value:", LT_max_value)
print("Minimum value:", LT_min_value)
```

```python
# Finding max and min values for FHR Hist Mean
hist_mean_max = df['FHR_hist_mean'].max()
hist_mean_min = df['FHR_hist_mean'].min()

print("Maximum value:", hist_mean_max)
print("Minimum value:", hist_mean_min)
```

```python
# Correlation matrix of dataframe
from matplotlib.colors import ListedColormap

matrix = df.corr()
plt.figure(figsize=(15,15))  

cmap = sns.diverging_palette(250, 10, s=80, l=55, n=9, as_cmap=True)

sns.heatmap(matrix, annot = True, cmap = 'Blues', center = 0)
```
