```python
# Import dependencies
import pandas as pd
from pathlib import Path
```

```python
# Preview CSV data
data = Path('../Resources/clean_fetal_health.csv')
fetal_df = pd.read_csv(data)
fetal_df.head()
```

```python
# Set 'A' column as index
fetal_df.reset_index(inplace=True)
```

```python
fetal_df.head()
```

```python
# Rename columns using a dictionary
new_column_names = {'index': 'Fetal ID'}
fetal_df = fetal_df.rename(columns = new_column_names)

fetal_df[0:60]
```

```python
fetal_df = fetal_df.drop(columns = 'fetal_health')
```

```python
fetal_df.head()
```

```python
fetal_df.to_json('fetal_data.json', orient='records') 
```
