```python
# Import dependencies
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Preview CSV data
data = Path('C:\\Users\\chere\\group10-final-project\\Resources\\clean_fetal_health.csv')
df = pd.read_csv(data)
df.head()
```

```python
# Total rows
total_rows = df.shape[0]
print(total_rows)
```

```python
# Finding max and min percentages (% Time with Abnormal Short-Term Variance) for binning
ST_max_value = df['abnorm_ST_Var_Perc'].max()
ST_min_value = df['abnorm_ST_Var_Perc'].min()

print("Maximum value:", ST_max_value)
print("Minimum value:", ST_min_value)
```

```python
# Finding max and min percentages (% Time with Abnormal Long-Term Variance) for binning
LT_max_value = df['abnorm_LT_Var_Perc'].max()
LT_min_value = df['abnorm_LT_Var_Perc'].min()

print("Maximum value:", LT_max_value)
print("Minimum value:", LT_min_value)
```

```python
# Finding max and min values for FHR Hist Mean
hist_mean_max = df['FHR_hist_mean'].max()
hist_mean_min = df['FHR_hist_mean'].min()

print("Maximum value:", hist_mean_max)
print("Minimum value:", hist_mean_min)
```

```python
# Correlation matrix of dataframe
from matplotlib.colors import ListedColormap

matrix = df.corr()
plt.figure(figsize=(15,15))  

cmap = sns.diverging_palette(250, 10, s=80, l=55, n=9, as_cmap=True)

sns.heatmap(matrix, annot = True, cmap = 'Blues', center = 0)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style("whitegrid")

# List of variables for individual distribution plots
variables = [
    'abnorm_ST_Var_Perc', 'abnorm_LT_Var_Perc', 'FHR_hist_mean', 
    'FHR_hist_median', 'mean_ST_Var', 'accelerations', 
    'FHR_hist_mode', 'mean_LT_Var', 'baseline_FHR_bpm', 'prolonged_decelerations'
]

# Define the figure size and grid
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(18, 24))
axes = axes.flatten()  # Flatten the grid for easy iteration

# Loop through the variables and create distribution plots
for i, var in enumerate(variables):
    sns.histplot(data=df, x=var, hue="fetal_health", multiple="stack", palette="viridis", ax=axes[i])
    axes[i].set_title(f'Distribution of {var} by Fetal Health', fontsize=14)
    axes[i].set_xlabel(var, fontsize=12)
    axes[i].set_ylabel('Count', fontsize=12)
    axes[i].legend(title='Fetal Health', title_fontsize='13', fontsize='12')

plt.tight_layout()
plt.show()
```
